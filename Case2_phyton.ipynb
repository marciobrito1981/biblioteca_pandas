```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP','Aplicativo')
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ','_')
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
# dropando todos os dados que possuem Nulos.
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
# preenchendo a coluna estado, onde tiver nulo, preencher com MS
vendas['estado'].fillna('MS',inplace=True)
vendas.info()
```

```python
#tirando a media de preços com a função mean
media_preco = vendas['Preço'].mean()

vendas['Preço'].fillna(media_preco, inplace = True)

vendas.info()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
# criando 2 bases novas, sendo 1 com os preços errados e outra com o preço correto
#preço errado
vendas_errado = vendas.query('Preço > Preço_com_frete')

#preço correto
vendas_correto = vendas.query('Preço < Preço_com_frete')
vendas_correto.head()
```

```python
vendas_correto.query('Nome_Departamento == "Esporte_e_Lazer" and estado == "SP"').describe()
```

```python
# Filtro de Departamento com mais vendas
vendas_correto.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending = False).reset_index()
```

```python
# Media de preço com frete  por nome de departamento, utilizando a função round, para deixar com 2 casas decimais
round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
```

```python
# Quantidade de vendas por data
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending=False)
```

```python
# Criando coluna para armazenar o mes
vendas_correto['Mes'] = vendas_correto['Data'].dt.month
vendas_correto.head()
```

```python
# Quantidade de vendas por mes
vendas_correto.groupby('Mes').idcompra.nunique().sort_values(ascending=False)
```

```python
# Importando a base de clientes

cliente = pd.read_excel('/content/cliente_varejo.xlsx')
cliente.head()
```

```python
# Analisando tabela de clienets
cliente[['idade','renda']].describe()
```

```python
# Transformando a coluna renda para float
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_correto.head()
```

```python
# Realizando o join entre as bases
vendas_cliente = vendas_correto.merge(cliente, how ='left', on='cliente_Log')
vendas_cliente.head()
```

```python
# Media de renda para cada canal de venda, utilizando o round, deixamos com 2 casas decimais e ordenando os dados do maior para o menor
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
# Media de clientes por idade, por bandeira.
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').reset_index(),2)
agg_idade_bandeira
```

```python
#Importando biblioteca de graficos
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (5,3)) # Tamanho do grafico
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green')
plt.ylabel("Media de Idade") #Alterando o nome do eixo Y
plt.title('Idade Média por bandeira', loc = 'center', fontsize = 18, fontweight='bold', color='black'); # Titulo do grafico
```

```python
fig = plt.figure(figsize = (5,3)) # Tamanho do grafico
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'blue')
plt.ylabel("Média de Renda") #Alterando o nome do eixo Y
plt.title('Renda Média por Venda', loc = 'center', fontsize = 18, fontweight='bold', color='black'); # Titulo do grafico
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

```python
#Gerando grafico de linha
fig = plt.figure(figsize=(10,4))

plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de Vendas')
x = vendas_por_data['Data']
y = vendas_por_data['idcompra']

plt.plot(x,y,  color = 'red', linewidth='2' )
plt.show()
```

```python
agg_dep_preco = round(vendas_correto.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(),2)
agg_dep_preco
```

```python
fig = plt.figure(figsize = (15,8)) # Tamanho do grafico
plt.bar(agg_dep_preco['Nome_Departamento'], agg_dep_preco['Preço_com_frete'], color = 'blue')
plt.ylabel("Preço Médio") #Alterando o nome do eixo Y
plt.title('Preco Médio por departamento', loc = 'center', fontsize = 18, fontweight='bold', color='black'); # Titulo do grafico
```
